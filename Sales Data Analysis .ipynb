### Data Queries

#### Overall Sales Trends

```python
# Finding the data for month_year based on the other_date
data['month_year'] = data['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Group data for month_year by sales
data_temp = data.groupby('month_year').sum()['sales'].reset_index()
```

```python
# Plot 
plt.figure(figsize=(20, 5))
plt.plot(data_temp['month_year'], data_temp['sales'], color='#39B7CD')

plt.xticks(rotation='vertical', size=10)
plt.show()
```

#### Top 10 products by sales

```python
# Group data for products by sales
product_sales = pd.DataFrame(data.groupby('product_name').sum()['sales'])
```

```python
# Sorts data in descending order
product_sales.sort_values(by=['sales'], inplace=True, ascending=False)
```

```python
# Top 10 products by sales
product_sales[:10]
```

#### Most selling products

```python
# Group the products based on quantity
best_selling_products = pd.DataFrame(data.groupby('product_name').sum()['quantity'])
```

```python
# Sorting the dataframe in descending order
best_selling_products.sort_values(by=['quantity'], inplace=True, ascending=False)
```

```python
# Top 10 most selling products
best_selling_products[:10]
```

#### Most preferred choice (mode/class) for shipping

```python
# Setting the figure size
plt.figure(figsize=(10, 8))

# Displays count of entried in each category using bars
sns.countplot(x='ship_mode', data=data)

# Display the figure
plt.show()
```

#### Most profitable category and sub-category

```python
# Group products by category and sub-category
category_subcategory = pd.DataFrame(data.groupby(['category', 'sub_category']).sum()['profit'])
```

```python
# Sort the values
category_subcategory.sort_values(['category','profit'], ascending=False)
```
